```python
#max hang, chi so axis 0 theo hang
print(np.apply_along_axis(max,0,a))
#min cot, chi so axis 1 theo cot
print(np.apply_along_axis(min,1,a))

#tong
def Sum(a):
    S=0
    for i in a:
        S +=i
    return S

print(np.apply_along_axis(Sum,1,a))

#tong cac chi so le

def Sum_odd(a):
    S = 0
    for i in range(0,len(a)):
        if i % 2 == 0:
            S += a[i]
    return S
#Sum cot
print(np.apply_along_axis(Sum_odd, 1, a))
#Sum hang
print(np.apply_along_axis(Sum_odd, 0, a))








```
