```python
def tinh_mean(sample):
    return sum(sample) / len(sample)

def tinh_median(sample):
    n = len(sample)
    index = n // 2
    if n % 2:
        return sorted(sample)[index]
    return sum(sorted(sample)[index - 1:index + 1]) / 2

def tinh_mode(sample):
    c = Counter(sample)
    return [k for k, v in c.items() if v == c.most_common(1)[0][1]]

t = []
for i in range(0,15):
    x = np.arange(1,11,1, dtype=int)
    t.append(x)
mat = np.matrix(t)
mat.shape

mean = np.apply_along_axis(tinh_mean,0,mat)
median = np.apply_along_axis(tinh_median,0,mat)
mode = np.apply_along_axis(tinh_mode,0,mat)
```
