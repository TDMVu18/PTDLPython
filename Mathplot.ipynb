```python
import numpy as np
import pandas as pd
import random
from sympy import *
import matplotlib.pyplot as plt
```

```python
#Dao ham
vec_x = np.linspace(-6,5,1500)
x = Symbol('x')
y = x*x*x + 3*x + 7
yprime = y.diff(x)
yprime_2 = yprime.diff(x)
print("Dao ham bac 1:",yprime)
print("Dao ham bac 2:",yprime_2)
print("Vecto x:",vec_x)
y = lambdify(x, y)
y1 = lambdify(x, yprime)
y2 = lambdify(x, yprime_2)
```

```python
y(vec_x)


```

```python
y1(vec_x)
```

```python
y2(vec_x)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(vec_x, y(vec_x))
ax.set_title("y(x)");
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(vec_x, y1(vec_x))
ax.set_title("y'(x)");
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(vec_x, y2(vec_x))
ax.set_title("y''(x)");
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(vec_x, y(vec_x))
ax.plot(vec_x, y1(vec_x))
ax.plot(vec_x, y2(vec_x))
ax.legend(["y(x)", "y'(x)", "y''(x)"])
plt.savefig("test.svg")
```
